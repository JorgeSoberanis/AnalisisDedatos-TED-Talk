### Pasamos ahora renombrar las columnas de ingles a español 
  Esto para poder manipularlos de una mejor manera y con referencias claras

```python
df_columns = [
    'titulo',
    'autor',
    'fecha',
    'visualisaciones',
    'likes',
    'link',
    'NAN6',
    'NAN7',
    'NAN8',
    'NAN9',
]
```

#### Asigno los nombres 

```python
datamuestra.columns = df_columns
```

#### Comprobamos si los datos fueron cambiados con exito

```python
print(datamuestra.info())
```

```python
alt.Chart(datamuestra).mark_bar().encode(
    x = alt.X('likes', axis = alt.Axis(labelAngle=0, labelFontSize=14)),
    y = alt.Y('visualisaciones', axis = alt.Axis(labelAngle=0, labelFontSize=14))
).properties(
    title = "Grafico de likes",
    width = 700,
)
```

```python
alt.Chart(datamuestra).mark_area(
    color = 'ligthblue',
    interpolate='step-after',
    line=True
).encode(
    x='visualisaciones',
    y= 'likes'
)
```
