```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_csv("TED.csv")
df.head()
```

#### Vamos a ordenar los datos de mayor a menor numero de likes: 

```python
orden_likes = df.sort_values("likes", ascending=False)
orden_likes.head()
```

#### Vamos a ordenar los datos de menor a mayor numero de likes

```python
orden_vistas = df.sort_values("likes", ascending=True)
orden_vistas.head()
```

#### Vamos a ordenar los autores de manera alfabetica:

```python
orden_autores = df.sort_values("author", ascending=True)
```

#### Sacamos datos random con el atributo sample

```python
datamuestra = df.sample(800, random_state =None)
datamuestra.head()
```

### Quiero renombrar las columnas de ingles a español 

```python
df_columns = [
    'titulo',
    'autor',
    'fecha',
    'visualisaciones',
    'likes',
    'link',
    'NAN6',
    'NAN7',
    'NAN8',
    'NAN9',
]
```

#### Comprobamos si los datos fueron cambiados con exito

```python
print(datamuestra.info())
```

```python
alt.Chart(datamuestra).mark_bar().encode(
    x = 'likes',
    y = 'visualisaciones',

    color= alt.condition(
        alt.datum.visualisaciones >= 1000000,
        alt.value('orange'),
        alt.value('steelblue'),
    )
)
```

```python
alt.Chart(datamuestra).mark_area(
    color = 'ligthblue',
    interpolate='step-after',
    line=True
).encode(
    x='visualisaciones',
    y= 'likes'
).interactive()
```
