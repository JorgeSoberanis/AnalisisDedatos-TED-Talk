```python
import pandas as pd
import numpy as np
import altair as alt
```

```python
df = pd.read_csv("TED.csv")
df.head()
```

#### Vamos a ordenar los datos de mayor a menor numero de likes: 

```python
orden_likes = df.sort_values("likes", ascending=False)
orden_likes.head()
```

#### Vamos a ordenar los datos de menor a mayor numero de likes

```python
orden_vistas = df.sort_values("likes", ascending=True)
orden_vistas.head()
```

#### Vamos a ordenar los autores de manera alfabetica:

```python
orden_autores = df.sort_values("author", ascending=True)
```

#### Sacamos datos random con el atributo sample

```python
datamuestra = df.sample(100, random_state =1000)
datamuestra.head()
```

### Quiero saber cuanta memoria me esta consumiendo esta proceso
